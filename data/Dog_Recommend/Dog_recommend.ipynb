```python
data = pd.read_csv("dogKor2.csv", encoding='utf8')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vector = CountVectorizer(ngram_range=(1,3))
```

```python
count_vector
```

```python
c_vector_keyword = count_vector.fit_transform(data['temper'])
```

```python
c_vector_keyword.shape
```

```python
c_vector_keyword.toarray() 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
specal_c_sim = cosine_similarity(c_vector_keyword, c_vector_keyword).argsort()[:,::-1]
```

```python
specal_c_sim.shape
```

```python
specal_c_sim
```

```python
def get_recommend_dog_list(df, dog_breed, top=10):
    # 특정 영화와 비슷한 영화를 추천해야 하기 때문에 '특정 영화' 정보를 뽑아낸다.
    target_dog_index = df[df['breed'] == dog_breed].index.values
    
    #코사인 유사도 중 비슷한 코사인 유사도를 가진 정보를 뽑아낸다.
    sim_index = specal_c_sim[target_dog_index, :top].reshape(-1)
    #본인을 제외
    sim_index = sim_index[sim_index != target_dog_index]

    #data frame으로 만들고 vote_count으로 정렬한 뒤 return
    result = df.iloc[sim_index][:10]
    return result
```

```python
get_recommend_dog_list(data, dog_breed="말티즈")
```
